- Bias is a measurement of how accurately a model can capture a pattern in a training dataset. If train error is big it is said to have a high bias
- low bias, low varaince - Balance fit
- low bias, high varaince - Overfit
- high bias, low varaince - Underfit



- ways to get balanced fit model is 
1. Cross Validation
2. Regularization
3. Dimensionality Reduction
4. Ensemble Techniques
