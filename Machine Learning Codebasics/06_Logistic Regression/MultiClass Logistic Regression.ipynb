```python
len(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
